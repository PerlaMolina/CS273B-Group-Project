# CPM normalization

```python
#sum all counts per sample
samplesums = adata_raw.sum(axis = 0)
samplesums
```

```python
#divide sum by 1,000,000 to derive scalefactor
scalefactor = samplesums/1000000
scalefactor
```

```python
#divide expression in each sample by corresponding scalefactor
adata_cpm = adata_raw.div(scalefactor, axis=1)   #axis=1 is important to divide per sample
adata_cpm
```

```python
#check that all samples summed expression == 1M
adata_cpm.sum(axis=0)
```

## Log2 normalize
(because its hella sparse and can't take log(0)

```python
#log2(CPM+1)
adata_norm = np.log2((adata_cpm + 1))
adata_norm
```
