# CPM normalization

```python
#sum all counts per sample
samplesums = adata_raw.sum(axis = 0)
samplesums
```

```python
#divide sum by 1,000,000 to derive scalefactor
scalefactor = samplesums/1000000
scalefactor
```

```python
#divide expression in each sample by corresponding scalefactor
adata_cpm = adata_raw.div(scalefactor, axis=1)   #axis=1 is important to divide per sample
adata_cpm
```

```python
#check that all samples summed expression == 1M
adata_cpm.sum(axis=0)
```

## Log2 normalize
(because its hella sparse and can't take log(0)

```python
#log2(CPM+1)
adata_norm = np.log2((adata_cpm + 1))
adata_norm
```

```python
adata_coord_reset = adata_coord.copy()
adata_coord_reset.columns = range(adata_coord_reset.shape[1])
adata_coord_reset.head()
```

```python
adata_cell_type_reset = adata_cell_type.copy()
adata_cell_type_reset.columns = range(adata_cell_type_reset.shape[1])
adata_cell_type_reset.head()
```

```python
adata_coord_with_ct = pd.concat([adata_coord_reset, adata_cell_type_reset])
adata_coord_with_ct
```

```python
adata_coord_with_ct_transpose = adata_coord_with_ct.T
adata_coord_with_ct_transpose
```

```python
adata_coord_with_ct_transpose.index = adata.obs['Donor ID']
adata_coord_with_ct_transpose
```

```python
# Create a dictionary to store the matrices for each patient
patient_matrices = {}

# Group by patient_id
grouped = adata_coord_with_ct_transpose.groupby('Donor ID')

# Iterate over each group
for patient_id, group in grouped:
    # Convert the group to a list of lists (matrix)
    matrix = group[['X', 'Y', 'Cell Type']].values.tolist()

    # Store the matrix in the dictionary
    patient_matrices[patient_id] = matrix
```
